# **Jak zdefiniować funkcję w Pythonie?**
W Pythonie funckję tworzy się komendą `def`. Następnie musimy podać nazwę naszej funkcji oraz w nawiasie wpisać ewentaualne `parametry`. 

**Przykład**


```python
def moja_funkcja(): #nazwę funkcji podajemy po komendzie def. Nawiasy i dwukropek po nazwie są konieczne.
  print("Moja pierwsza funkcja")
```

`Wcięcie` to wiodąca spacja (spacje i tabulatory) przed jakąkolwiek instrukcją w Pythonie. Python traktuje instrukcje, które mają ten sam poziom wcięć jako pojedynczy blok kodu. Wcięcia zwiększają również czytelność kodu.

# **Parametry i argumenty funkcji**

Informacje mogą być przekazywane do funkcji jako `argumenty`.

`Argumenty` są podane po nazwie funkcji, w nawiasach. Możesz dodać dowolną liczbę argumentów, po prostu oddziel je przecinkiem.

Poniższy przykład ma funkcję z jednym argumentem `x`. Aby ją wywowłać, musimy podać w nawiasię liczbę, która ma pełnić rolę `x`. Jeśli więc podamy jako argument liczbę `5`, to `fajna_funkcja` wyświetli napis `Funkcję są fajne!` 5 razy.


```python
def fajna_funkcja(x):
  print('Funkcje są fajne! '*x)
```

```python
fajna_funkcja(5)
```

`Funkcje` mogą, ale nie muszą, brać argumentów i mogą, ale nie muszą dawać wyniku.
`Parametr` to zmienna wymieniona w nawiasach w definicji funkcji. W naszym przypadku to po prostu `x`.

Jaka jest różnica pomiędzy parametrem, a argumentem funkcji?

Ogólnie rzecz biorąc, kiedy ludzie mówią `parametr/argument`, mają na myśli to samo, ale główna różnica między nimi polega na tym, że parametr jest tym, co jest zadeklarowane w funkcji, podczas gdy argument jest tym, co jest przekazywane podczas wywoływania funkcji. `Return` mówi funkcji, że ma zwrócić wartość (w tym przypadku a+b).

```python
def add(a, b):
    return a+b

add(5, 4)
```

Tutaj `parametry` to `a` i `b`, a przekazywane `argumenty` to `5` i `4`.

# **Przykład 2**
Napisz program, który zwraca liczbę miejsc zerowych funkcji kwadratowej zadanej przez użytkownika. 

Rozwiązanie:

Przypomnijmy, że o ilości miejsc zerowych funkcji kwadratowej decyduje wartość wyróżnika (delty).

Wzór na deltę:

$Δ = b^2 - 4ac$

*   jeśli $Δ > 0$ są dwa miejsca zerowe,
*   jeśli $Δ = 0$ istnieje jedno miejsce zerowe,
*   jeśli $Δ < 0$ funkcja nie ma miejsc zerowych.

Wystarczy napisać program, które sprawdza te warunki dla danych liczb $a,b,c ∈ ℜ$ (współczynników funkcji kwadratowej), gdzie $a ≠ 0$.



```python
def funkcja_delta(a,b,c):
  delta = b**2 - 4*a*c

  if delta > 0: 
    print('Funkcja ma dwa miejsca zerowe')
  if delta == 0: 
    print('Funkcja ma jedno miejsca zerowe') # zauważ że mamy == zamiast =
  if delta <0: 
    print('Funkcja nie ma miejsc zerowych')
```

# **Sprawdźmy działanie programu**

Mamy daną funckję kwadratową $f(x) = 2x^2 + 4x -7$

$a = 2, b = 4, c = -7$

```python
funkcja_delta(2,4,-7)
```

# **Przykład 3.**
Stwórz funkcję `funkcja_klamrowa` który oblicza wartość funkcji `f(x)` danej wzorem:

$$ f(x) =   \left\{
\begin{array}{ll}
      -2 & x<-2 \\
      2x +2 & x \in[-2,2) \\
      x^3 -2 & x \geq 2 \\
\end{array} 
\right.  $$

Rozwiązanie:

```python
def funkcja_klamrowa(x):
  if x<-2:
    wynik = -2
  elif x>=-2 and x<2:  #ważna linijka
    wynik = 2*x +2
  else:
    wynik = x**3 - 2
  return wynik
```

```python
funkcja_klamrowa(3)
```

# **Przypomnienie: Instrukcje Warunkowe**

Instrukcja `if (...) else` ocenia wyrażenie testowe i wykonuje treść `if` tylko wtedy, gdy warunek testowy ma wartość `True`.

Jeśli warunek testowy ma wartość `False`, wykonywana jest treść `else`.

`elif` jest skrótem od `else if`. Dzięki `elif` możemy sprawdzać wiele warunków testowych.

Jeśli warunek testowy dla `if` jest `False`, sprawdzany jest warunek następnego bloku `elif` i tak dalej.

Jeśli wszystkie warunki testowe są `False`, wykonywana jest treść `else`.

`Pamiętaj!` Jeśli twój program ma sprawdzać wiele warunków testowych to pierwszy sprawdzaj komendą `if` ale kolejne komendą `elif`. Na koniec daj `else` aby program wiedział co zrobić w przypadku, gdy żadne warunki nie będą spełnione.

Przykład 1 można było więc rozwiązać również w taki sposób (lepszy):

```python
def funkcja_delta(a,b,c):
  delta = b**2 - 4*a*c

  if delta > 0:   #warunek testowy

    print('Funkcja ma dwa miejsca zerowe') #treść

  elif delta == 0: #kolejny warunek testowy

    print('Funkcja ma jedno miejsca zerowe') # treść 

  else: 

    print('Funkcja nie ma miejsc zerowych') #treść
```
